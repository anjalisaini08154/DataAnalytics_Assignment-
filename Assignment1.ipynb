Ques1. Why are dictionary keys required to be immutable in Python? Explain with a small example.

Ans:
In Python, dictionary keys must be immutable because dictionaries work using a concept called hashing. Each key is converted into a unique hash value, and this value is used to store and retrieve the data quickly.
If a key is mutable (like a list), it can change after being added to the dictionary. When the value of the key changes, its hash value will also change. This creates a problem because Python will not be able to find the correct location of that key in the dictionary.
Therefore, keys must be immutable so that their hash value remains constant and data can be accessed correctly


```python
# Correct example (immutable key)
d = {"name": "Anjali"}
print(d["name"])
```

•	Here "name" is a string (immutable), so it works fine.

```python
# Wrong example (mutable key)
d = {[1, 2]: "data"}
```

Ques2. Write a program that takes a list of elements and creates a dictionary showing the frequency of each element.

Ans:


```python
# Input list
lst = [1, 2, 2, 3, 1, 4, 2]

# Empty dictionary
freq = {}

# Counting frequency
for item in lst:
    if item in freq:
        freq[item] += 1
    else:
        freq[item] = 1
print(freq)
```

Ques4. Write a program to merge two dictionaries and resolve common keys by keeping the larger value.

Ans:


```python
#Input dictionaries
d1 = {"a": 10, "b": 20, "c": 30}
d2 = {"b": 25, "c": 15, "d": 40}

# Merged dictionary
result = {}

# Add elements from d1
for key in d1:
    result[key] = d1[key]

# Merge d2 and compare values
for key in d2:
    if key in result:
        result[key] = max(result[key], d2[key])  # keep larger value
    else:
        result[key] = d2[key]

print(result)
```

Ques5. Using a dictionary comprehension, create a dictionary from a list of integers where keys are numbers and values are their cubes, but only for odd numbers.

Ans:


```python
# Input list
lst = [1, 2, 3, 4, 5, 6]

# Dictionary comprehension (only odd numbers)
result = {x: x**3 for x in lst if x % 2 != 0}

print(result)
```
